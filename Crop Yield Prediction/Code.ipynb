```python
def undummy(df):
    category_cols = [col for col in df.columns if col.startswith('Soil_type:_')]
    df['Soil_Type'] = df[category_cols].idxmax(axis=1).str.replace('Soil_type:_', '').astype(int)
    df = df.drop(columns=category_cols)
    return df
```

```python
import pandas as pd

# read the CSV file into a data frame
data = pd.read_csv('Final_Data.csv')

# extract the desired columns into a new data frame
columns_to_extract = ['Crop:_Bajra', 'Crop:_Wheat', 'Crop:_Jowar']
new_data = data[columns_to_extract]
```

```python
import matplotlib.pyplot as plt

# create a dictionary to hold the values for each column
values = {}
for col in columns_to_extract:
    values[col] = new_data[col].sum()

# plot a pie chart of the values
fig, ax = plt.subplots()
ax.pie(values.values(), labels=values.keys(), autopct='%1.1f%%', startangle=90)
ax.set_title('Data Distribution')

plt.show()
```

```python
import matplotlib.pyplot as plt

# extract the values for each column
crop_bajra_values = new_data['Crop:_Bajra']
crop_jowar_values = new_data['Crop:_Jowar']
crop_wheat_values = new_data['Crop:_Wheat']

# set the positions of the bars on the x-axis
bar_positions = [1, 2, 3]

# plot the bars for each column
fig, ax = plt.subplots()
ax.bar(bar_positions, [crop_bajra_values.sum(), crop_jowar_values.sum(), crop_wheat_values.sum()])
ax.set_xticks(bar_positions)
ax.set_xticklabels(['Crop:_Bajra', 'Crop:_Jowar', 'Crop:_Wheat'])
ax.set_ylabel('Values')
ax.set_title('Data Distribution')

plt.show()
```
